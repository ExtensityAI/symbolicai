```python
x = ai.Type('This function transform natural language text to python code.', static_context='You are an SQL expression....')
x.add(...) != ai.Symbol(x).add(...) # other function....
x(...) # forward behavior

x = ai.Function('This function transform natural language text to python code.') # just the function
x.add(...) == ai.Symbol(x).add(...)
x(...) # other function....

y = ai.Type('Classify numbers, only give even or odd as an answer.', examples=["if '5' then 'odd' else 'even'"])
z = ai.Type('This function transform natural language text to python code.')
f = ai.Fork()
```

```python
x.compose()
x.summarize()
x()
```

```python
w = f(Sequence(
        x,
        y
    ))

Sequence(
    w,
    y, 
    w
)
```

```python
x.ismember # False
```
